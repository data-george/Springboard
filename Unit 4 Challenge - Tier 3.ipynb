**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(two_decades, BOROUGH):
    
    #Use filtered data frame(two_decades) and filter rows by specific borough
    two_decades_borough = two_decades[two_decades['London_Boroughs']== BOROUGH]
    
    #Extract 1998 mean price
    BOROUGH_1998 = two_decades_borough.loc[1998,:]
    BOROUGH_1998_AVG_PRICE = BOROUGH_1998['Average_Price'].mean()
    
    #Extract 2018 mean price
    BOROUGH_2018 = two_decades_borough.loc[2018,:]
    BOROUGH_2018_AVG_PRICE = BOROUGH_2018['Average_Price'].mean()
    
    #calculate price ratio between 1998 and 2018
    RATIO = BOROUGH_1998_AVG_PRICE/BOROUGH_2018_AVG_PRICE
    
    return RATIO
   
   
```

```python
#Make a list of the unique Boroughs
Boroughs = list(set(two_decades['London_Boroughs']))
Boroughs
```

Let's get the actual list of boroughs from the link provided above and here once more [(here's some info for the curious)](https://en.wikipedia.org/wiki/London_boroughs)

```python
#This list of the actual borroughs will be used to filter our list and make sure we use 
#just the 32 official borroughs and also that the borrough names are  
#correct in our analysis, so we can simply reply to the question regarding
#which burroughs had the highest increase in average household cost. 
#You can see that we have 45 entries, which means some of them are not
#actual burroughs

thirty_two = ['Camden',
'Greenwich',
'Hackney',
'Hammersmith & Fulham', #slightly modified to match our dataset see link above (just Hammersmith)
'Islington',
'Kensington & Chelsea', #slightly modified -- just ampersand instead of 'and'
'Lambeth',
'Lewisham',
'Southwark',
'Tower Hamlets',
'Wandsworth',
'Westminster',
'Barking & Dagenham', #slightly modified to match our dataset see link above (just Barking)
'Barnet',
'Bexley',
'Brent',
'Bromley',
'Croydon',
'Ealing',
'Enfield',
'Haringey',
'Harrow',
'Havering',
'Hillingdon',
'Hounslow',
'Kingston upon Thames',
'Merton',
'Newham',
'Redbridge',
'Richmond upon thames',#slightly modified to match our dataset - see link above (just Richmond)
'Sutton',
'Waltham Forest'] 
#note that there is a table near the top of our link and another list of boroughs below that, this
#list actually very closely matches the second list on the page (probably wouldn't have needed edits)
thirty_two = [b.upper() for b in thirty_two]
thirty_two
```

```python
len(thirty_two)
```

```python
#let's see if the final set results in 32 entries
Boroughs = set(thirty_two).intersection(set(Boroughs))
```

```python
len(Boroughs)
```

```python
#Use the create_price_ratio function on each of the boroughs on the list of unique boroughs
RATIOS = [create_price_ratio(two_decades,BOROUGH) for BOROUGH in Boroughs] 
```

```python
#Create a dictionary key: name of borough, value: price ratio
ratio_by_borough = {BOROUGH:RATIO for BOROUGH, RATIO in zip(Boroughs, RATIOS)}
```

```python
ratio_by_borough
```

```python
#calculate greatest average increase, this should be the minimum value in the dictionary
#that would mean the difference is greater (the denominator of the ratio is larger)
greatest_average_increase = min(ratio_by_borough.keys(), key=(lambda k: ratio_by_borough[k]))
greatest_average_increase
```

```python
#Let's check if this makes sense by plotting the average household price change in Hackney
two_decades = df.loc[1998:2018, :]
HACKNEY_TWO = two_decades[two_decades['London_Boroughs']=='HACKNEY']
HACKNEY_TWO.plot(title='HACKNEY AVERAGE HOUSEHOLD PRICE CHANGE')
```
